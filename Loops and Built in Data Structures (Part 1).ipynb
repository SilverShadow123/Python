For Loop

```python
for i in range(10):
    print(i+1,end=' ')
print('\n')
for j in range(2,7):
    print(j,end=' ')
print('\n')
for k in range(10,1,-1):
    print(k,end=' ')
print('\n')
for l in range(10,1,-2):
    print(l,end=' ')
```

Nested Loop

```python
# one loop inside another loop
for i in range(5):
    for j in range(5):
        print(i,j)
```

Break

```python
for i in range(2,20,2):
    if i == 5:
        break
    print(i)
print('i am out of loop')


```

Continue

```python
for i in range(1,20):
        print(i)
        if i%2 ==0:
          continue
```

List Comprehention

```python
squares =[i*i for i in range(10)]
print(squares)
```

```python
even_square=[i*i for i in range(10) if i%2==0]
print(even_square)
```
