```python
number = 80
if number>40:
    print('passed')
else:
    print('failed')

if number > 80:
    print('Letter Grade A')
elif number > 70:
    print('Letter Grade B')
elif number > 60:
    print('Letter Grade C')
elif number > 50:
    print('Letter Grade D')
else:
    print('Letter Grade F')
```

Logical Operators

```python
age = 18
salary = 5100
if age > 18 and salary > 5000:
    print('Eligible for loan')
else:
    print('Not Eligible for loan')

if age > 18 or salary > 5000:
    print('Eligible for loan')
else:
    print('Not Eligible for loan')

if not age > 18:
    print('Not Eligible for loan')
else:
    print('Eligible for loan')
```
