Tuple

```python
tpl = (1,2,3)
print(type(tpl))

a_list = [1,2,3,4,5]
print(type(a_list))

a_tuple = tuple(a_list)
print(type(a_tuple))
```

Sets

```python
a_set = {1,1,2,3,5,4,1,6,7,8}
print(a_set)
```

Sets Method

```python
a_set = {1,1,2,3,5,4,1,6,7,8}

a_set.add(9)
print(a_set)

a_set.remove(9)
print(a_set)

a_set.pop()
print(a_set)

a_set.discard(9)
print(a_set)


a_set.clear()
print(a_set)

a_set = {1,2,3,4,5}
b_set = {4,5,6,7,8}

c_set = {1,2,3}
d_set = {1,2,3,4}

print(a_set.union(b_set))
print(a_set.intersection(b_set))
print(a_set.difference(b_set))
print(a_set.symmetric_difference(b_set))
print(c_set.issubset(d_set))
print(c_set.issuperset(d_set))
```

Dictionaries

```python
a_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
print(a_dict)
print(type(a_dict))
print(a_dict['name'])

stu_a_dict ={
    (1,2): 'Seaum',
    (3,4): 'John',
    (5,6): 'Doe'
}

print(stu_a_dict)
print(type(stu_a_dict))
print(stu_a_dict[(1,2)])
```

Dictionaries Methods

```python
a_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
print(a_dict.keys())
print(a_dict.values())
print(a_dict.items())
print(a_dict.get('name'))
print(a_dict.pop('age'))
print(a_dict.popitem())
print(a_dict)
print(a_dict.update({'city': 'Los Angeles'}))
print(a_dict)
print(a_dict.clear())
print(a_dict.copy())
print(a_dict.fromkeys(['name', 'age', 'city'], 0))
print(a_dict.setdefault('name', 'John'))
print(a_dict)
```
