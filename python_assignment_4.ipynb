```python
def process_scores(scores, threshold):
  above =[]
  below_or_equal=[]
  for i in scores:
    if i > threshold:
      above.append(i)
    else:
      below_or_equal.append(i)
  return above, below_or_equal

scores =[1, 3, 4,5,6,7,2]
threshold =3
above, below_or_equal = process_scores(scores, threshold)

print('Scores above threshold:', above)
print('Scores below or equal:', below_or_equal)
```

```python
def convert_temp(value, to_unit='C'):
  if to_unit == 'C':
    return (5/9) * (value - 32)
  elif to_unit == 'F':
    return (9/5) * value + 32
  else:
    return 'Invalid Unit'

fahrenheit_to_celsius = convert_temp(98)
celsius_to_fahrenheit = convert_temp(36, to_unit='F')
print(f'Fahrenheit to Celsius: {fahrenheit_to_celsius:.2f}째C')
print(f'Celsius to Fahrenheit: {celsius_to_fahrenheit:.2f}째F')
```

```python
x = 100  # Global variable

def demonstrate_scope():
    # First, print the global value using globals()
    print(f"Reading global x: {globals()['x']}")

    # Now create a local variable with the same name
    x = 500  # This creates a LOCAL variable
    print(f"Local x: {x}")

demonstrate_scope()

print(f"Global x: {x}")
```

```python
def safe_divide(a,b):
  try:
    result = a/b
    return {
        'success': True,
        'result': result
    }
  except ZeroDivisionError:
    return {
        'success': False,
        'error': 'Division by zero is not allowed'
    }
  except TypeError:
    return {
        'success': False,
        'error': 'The input you have given is invalid'
    }

print(safe_divide(8,4))
print(safe_divide(3,0))
print(safe_divide('a','b'))
```

```python
class BankAccount:
  def __init__(self, owner, balance):
    self.owner = owner
    self.balance= balance
  def deposit(self, amount):
    if amount > 0:
      self.balance += amount
      print(f'Deposited : ${amount:.2f}')
    else:
      print(f'Deposited Amount Must Be Positive')
  def withdraw(self, amount):
    if self.balance < amount:
      print('Insufficient Funds! Withdrawal Canceled.')
    elif amount <=0:
      print('Withdrawal Amount Must be Positive.')
    else:
      self.balance -= amount
      print(f'Withdrawn: ${amount:.2f}')
  def get_balance(self):
    print(f'\nAccount Name: {self.owner}\nBalance: ${self.balance:.2f}\n')

account = BankAccount('Seaum', 1000)

account.deposit(50)
account.get_balance()
account.withdraw(30)
account.withdraw(1150)
account.deposit(300)
account.get_balance()

```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def average(self):
        total = sum(self.marks.values())
        return total / len(self.marks)


student1 = Student(
    "Seaum",
    {"Math": 85, "Physics": 78, "Chemistry": 90}
)

student2 = Student(
    "Arafat",
    {"Math": 72, "English": 88, "Biology": 80}
)

print(f"{student1.name}'s average score: {student1.average():.2f}")
print(f"{student2.name}'s average score: {student2.average():.2f}")
```

```python
import os
from datetime import datetime


class InventoryReport:
    def __init__(self, item_name, stock):
        self.item_name = item_name
        self.stock = stock


def generate_inventory_summary(report_obj, filename="inventory_report.txt", header="INVENTORY SUMMARY"):
    try:
        if not isinstance(report_obj, InventoryReport):
            raise TypeError("Invalid object type. Expected InventoryReport.")

        total_stock = sum(report_obj.stock.values())

        current_time = datetime.now().strftime("%d-%m-%Y %I:%M:%S %p")

        with open(filename, "w") as file:
            file.write(f"{header}\n")
            file.write(f"Generated on: {current_time}\n\n")
            file.write(f"Item Name: {report_obj.item_name}\n")
            file.write("Stock Details:\n")

            for location, count in report_obj.stock.items():
                file.write(f"  {location}: {count}\n")

            file.write(f"\nTotal Stock: {total_stock}\n")

        return os.path.abspath(filename)

    except Exception as e:
        print(f"Error: {e}")


inventory = InventoryReport(
    "Laptop",
    {
        "Warehouse A": 50,
        "Warehouse B": 30,
        "Store Front": 20
    }
)

path1 = generate_inventory_summary(inventory, "report1.txt", "WAREHOUSE REPORT")
print("Report saved at:", path1)

path2 = generate_inventory_summary(
    report_obj=inventory,
    filename="report2.txt",
    header="INVENTORY STATUS"
)
print("Report saved at:", path2)

generate_inventory_summary("NotAnInventoryObject")
```
