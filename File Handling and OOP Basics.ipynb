Reading and Writing Files in Pythons

```python
file = open('test.txt', 'w')
file.write('Hello World')
file.close()
file = open('test.txt', 'r')
print(file.read())
file.close()
file = open('test.txt', 'a')
file.write('Hello World\n')
file.close()
```

```python
file = open('test.txt', 'r+')
file.write('Dragon')
file.close()
file = open('test.txt', 'r')
print(file.read())
file.close()
```

```python
file = open('test.txt', 'w+')
file.write('Hello World\n')
file.write('Dragon bll z')
file.close()
file = open('test.txt', 'r')
print(file.read())
file.close()
```

```python
file = open('test.txt', 'r')
lines= file.readline()
print(lines)
file.close()
```

With Statement

```python
with open('note.txt', 'w') as file:
    file.write('Hello World')
```

```python
with open('note.txt', 'r') as file:
    print(file.read())
```

Introduction to OOP

```python
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color
my_car = Car('Toyota', 'Red')
print(my_car.brand)
print(my_car.color)
```

```python
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color
    def drive(self):
        print(f'The {self.brand} car is driving')
my_car = Car('Toyota', 'Red')
my_car.drive()
```

```python
class Student:
    def __init__(self, name, id, gender, subject):
        self.name = name
        self.id = id
        self.gender = gender
        self.subject = subject
    def study(self):
        print(f'{self.name} is studying {self.subject}')
    def display(self):
        print(f'Name: {self.name}')
        print(f'ID: {self.id}')
        print(f'Gender: {self.gender}')
        print(f'Subject: {self.subject}')
student = Student('John', '123', 'Male', 'Math')
student.study()
student.display()

```

Mini Project: Student Grade Calculator

```python
class Student:
    def __init__(self, name, marks):
        self.name =name
        self.marks = marks
    def calculate_grade(self):
        if self.marks >= 90:
            return 'A'
        elif self.marks >= 80:
            return 'B'
        elif self.marks >= 70:
            return 'C'
        elif self.marks >= 60:
            return 'D'
        else:
            return 'F'
    def display(self):
        print(f'Name: {self.name}')
        print(f'Marks: {self.marks}')
        print(f'Grade: {self.calculate_grade()}')
name = input('Enter name: ')
marks = int(input('Enter marks: '))
student = Student(name, marks)
student.display()
with open('grade.txt', 'a') as file:
    file.write(f'Name: {student.name}\n')
    file.write(f'Marks: {student.marks}\n')
    file.write(f'Grade: {student.calculate_grade()}\n')
print('Grade saved to grade.txt')
with open('grade.txt', 'r') as file:
    print(file.read())
```
