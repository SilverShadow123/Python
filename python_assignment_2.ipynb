```python
name = input('Enter Your Full Name: ')
print('Title Case:', name.title())
print('First Character:', name[0])
print('Last Character:', name[-1])
print('Reversed Name:', name[::-1])
print("Count of 'a':", name.lower().count('a'))
print('Without Spaces:', name.replace(' ',''))
```

NUMBERS, MATH & OPERATOR PRECEDENCE

```python
number = int(input('Enter a Number: '))
if number>=1:
  print(f'{number} is Positive')
else:
  print(f'{number} is Negative')

if number % 2 == 0:
  print(f'{number} is even')
else:
  print(f'{number} is odd')
```

```python
subject_1=float(input('Enter Subject 1 mark: '))
subject_2=float(input('Enter Subject 2 mark: '))
subject_3=float(input('Enter Subject 3 mark: '))

total = subject_1+subject_2+subject_3
average = total/3

print('Total :', total)
print('Average:', format(average, '.2f'))

if subject_1 >= 40 and subject_2 >=40 and subject_3>=40:
  print('Pass')
else:
  print('Fail')

if average >= 75:
  print('Distinction')
else:
  print('Normal Pass')
```

LOOPS

```python
num = 1

while num <= 50:
    if num == 37:
        break
    if num % 4 == 0:
        num += 1
        continue

    print(num)
    num += 1
```

```python
number = int(input('Enter a Number to Print its Multiplication Table: '))

print(f'\nMultiplication Table of {number}:')

for i in range(1,13):
  print(f'{number} X {i} = {number * i}')
```

LISTS & LIST COMPREHENSION

```python
nums =[12, 4, 8, 1, 9, 3, 4, 4]
nums.sort
print('Sort:', nums)
nums.insert(2,100)
print('Inserting 100 at index 2:', nums)
nums.remove(1)
print('Removing the value 1:',nums)
print('Max Value:', max(nums))
print('Min Value:', min(nums))
nums.reverse()
print('Reverse:', nums)
print('Counting how many times 4 appears:', nums.count(4))
```

```python
integers_list =[12,32,5,75,86,97,2,63,10,4]

print('Interger List:', integers_list)
odd_integers = [i for i in integers_list if i%2!=0]
print('Odd Number:', odd_integers)
greter_then_10 =[i for i in integers_list if i > 20]
print('Numbers greter than 20', greter_then_10)
less_then_10 = [i if i > 10 else 0 for i in integers_list]
print('Replaced all Numbers less then 10 with 0:', less_then_10)
```

```python
l = []
for i in range(1,101):
  l.append(i)
print('List:', l)
div_by_3_not_5 = [i for i in l if i % 3 == 0 and i % 5 != 0]
print('After divisibleb by 3 but not divisible by 5:', div_by_3_not_5)
```
