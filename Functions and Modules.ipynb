Defining Functions

```python
def greet():
    print('Hello World')

greet()

```

Parameters and Arguments

```python
def greet(name):
    print('Hello',name)
name = input('Enter name ')
greet(name)
```

```python
def greet(name, age):
    print(f"Hello {name} You are {age} years old")
name = input('Enter name ')
age = input('Enter age ')
greet(name, age)
```

Return Values

```python
def add(x, y):
    return x + y
add(5, 3)
```

```python
def even_odd_detector (number):
    if number % 2 == 0:
        return 'Even', True
    else:
        return 'Odd', False
even_odd_detector(5)
print(type(even_odd_detector(5)))
```

```python
even_odd_detector(5)[0]
```

```python
s, b = even_odd_detector(5)
print(s)
print(b)
```

Default and keyword arguments

```python
def greet(name='World'):
    print('Hello',name)
greet()
greet('Seaum')
```

```python
def add(a=1, b=1):
    return a+b
add()
add(5, 3)
```

```python
def exp(a=1, b=1):
    return a**b
print(exp())
exp(b=2, a=3)
```

Variable Scope

```python
x=20
def f():
    global x
    x =10
    print(x-10)
    return x+10
f()
print(f())
print('done')
print(x)
```

Try and Except

```python
try:
    int('Hello World')
except:
    print('Invalid Input')
```

```python
try:
    num = int(input('Enter a number: '))
    result = 10/num
    print(result)
except ValueError:
    print('Enter a valid number')
except ZeroDivisionError:
    print('Cannot divide by zero')
```
