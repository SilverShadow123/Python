Section C

```python
sales = {
    "Jan": {"total": 1200, "returns": 2},
    "Feb": {"total": 1500, "returns": 3},
    "Mar": {"total": 1700, "returns": 1}
}

for month, data in sales.items():
    data['total'] = round(data['total'] * 1.05, 2)
print("After 5% increase:", sales)

sales.update({"Apr": {"total": 1600, "returns": 2}})
print("After adding Apr:", sales)

max_returns = max(data['returns'] for data in sales.values())

for month in list(sales.keys()):
    if sales[month]['returns'] == max_returns:
        del sales[month]

print("After deleting highest returns:", sales)

above_1500 = []
for month, data in sales.items():
    if data['total'] > 1500:
        above_1500.append(month)

print("Months with total > 1500:", above_1500)
```

```python
students = {
    1: {"name": "Riya", "marks": {"math": 88, "eng": 76, "sci": 90}},

    2: {"name": "Fahim", "marks": {"math": 92, "eng": 81, "sci": 79}},

    3: {"name": "Simran", "marks": {"math": 67, "eng": 72, "sci": 69}}
}
for id, mark in students.items():
     mark['marks'].update({'computer':85})
print('Adding computer:85 to every students marks: ',students)

for id, data in students.items():
    data['total'] = sum(data['marks'].values())
print('After Calculating each total and adding: ',students)

max_total = max(data['total'] for data in students.values())

for id, data in students.items():
    if data['total'] == max_total:
      print('Student with the highest total:',data['name'])

print('Above 70 Marks: ', end='')
for id, data in students.items():
  for mark in data['marks'].values():
    if mark>70:
      print(data['name'],end=', ')
      break
```

```python
inventory = {
    "fruits": {"apple": 50, "banana": 30},

    "vegetables": {"carrot": 40, "spinach": 25},

    "drinks": {"water": 100, "juice": 40}
}


for category, items in inventory.items():
  for item, qty in items.items():
     items[item] = int(qty*1.10)

print('Updated inventory with 10% increase:',inventory)

high_qty_inventory = {}

for category, items in inventory.items():
    filtered_items = {item: qty for item, qty in items.items() if qty > 40}
    if filtered_items:
        high_qty_inventory[category] = filtered_items

print('Items whose updated quantity > 40:',high_qty_inventory)
```
