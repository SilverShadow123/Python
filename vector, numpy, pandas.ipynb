```python
import timeit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from seaborn.external.husl import dot_product

feature_vector = np.array([5.1, 3.5, 1.4, 0.2, 5.0])

print(f'shape: {feature_vector.shape}')

data_matrix = np.array([[10, 20.0, 30],[40, 50, 60]])

print(f'Shape: {data_matrix.shape}')
```

```python
large_array = np.random.rand(100000)
python_list = large_array.tolist()

np_time = timeit.timeit('large_array * 2', globals={'large_array':large_array}, number=10)

def list_op(data):
    return [x *2 for x in data]
list_time = timeit.timeit('list_op(python_list)', globals={'list_op':list_op, 'python_list': python_list}, number=10)

print(f'Python: {list_time:.5f}')
print(f'Numpy: {np_time:.5f}')
print(f'NumPy is {list_time/np_time:.0f}x faster')


```
