Simple Calculator

```python
#Type casting both input as float and taking numbers from user
num1 = float(input('Enter num 1: '))
num2 = float(input('Enter num 2: '))

#Printing Addition, Subtraction, Divesion, Multiplication
print('\nCalculation: ')
print('Addition : ',num1+num2)
print('Subtraction : ',num1-num2)
print('Division : ',num1/num2)
print('Multiplication : ',num1*num2)
```

User Introduction Program

```python
#Taking name, city, favourite_hobby as input from user
name = input('Enter Your Name: ')
city = input('Enter Your City: ')
favourite_hobby = input('Enter Your Favourite Hobby: ')

#Using f'' string and printing the reasult
print(f"\nHello {name}! You live in {city} and you enjoy {favourite_hobby}.")
```

Temperature Converter

```python
#Taking c as calcius as input
c=float(input("Enter Celcius: "))

#celcius to fahrenheit formula
f=(c*9/5)+32

#printing the converted fahrenheit result
print(f"\nTemperature in Fahrenheit is: {f}°F")
```

Billing Program

```python
#Taking item_name as Item name, item_price as Item price, item_quantity as Quantity from the user
item_name= input('Enter Item Name: ')
item_price=float(input('Enter Item Price: '))
item_quantity=int(input('Enter Quantity: '))

#arithmatic calculation
total = item_price * item_quantity

#printing the reasult and the item name
print(f"\nTotal Bill for {item_name} = {total}")
```
